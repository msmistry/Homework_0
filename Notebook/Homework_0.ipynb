```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Salaries.csv', index_col='playerID')
data
```

```python
data.head()
```

```python

# Filter players who are registered in ATL or HOU and have a salary greater than one million
filtered_players = data[(data["teamID"] .isin(["ATL", "HOU"])) & (data["salary"] > 1_000_000)]

# Select only the player IDs
player_ids = filtered_players.index

print(player_ids)
```

```python
data.describe()
```

```python
atl_salaries = data[data["teamID"] == "ATL"]["salary"]

stats = atl_salaries.describe()
stats
```

```python
data_dict = {col: [] for col in data.columns}

for _, row in data.iterrows():
    for col in data.columns:
        data_dict[col].append(row[col])

print(data_dict)
```
